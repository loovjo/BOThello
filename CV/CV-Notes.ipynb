```python
import numpy as np
import scipy as sp
from scipy import ndimage as ni
import skimage as si
import skimage.feature
import matplotlib.pyplot as plt
```

```python
def show(img):
    plt.imshow(img)
    plt.show()
```

```python
im = sp.misc.imread("dataset/game0/img0.jpg")[:,16:-16,0] / 256
show(im)
```

```python
sx = ni.sobel(im, axis=0, mode="constant")
sy = ni.sobel(im, axis=1, mode="constant")
s = np.hypot(sx, sy)
show(s)
```

```python
show(skimage.feature.canny(im))
```
